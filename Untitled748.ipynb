```python
# Replace PSO with QPSO
# All outputs saved with qpso_ prefix

import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ==============================
# PATHS
# ==============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Export Demand Forecasting System"
CSV_PATH = os.path.join(BASE_DIR, "Table_B-4.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv(CSV_PATH)

# ==============================
# WIDE TO LONG FORMAT
# ==============================

rows = []

for _, row in df.iterrows():

    market = row["Market"]
    parameter = row["Parameter"]

    for col in df.columns:

        if "Q/V" in col:

            year_label = col.split(" - ")[0]
            percent_col = f"{year_label} - %"

            rows.append({
                "Market": market,
                "Parameter": parameter,
                "Year_Label": year_label,
                "Year": int(year_label.split("-")[0]),
                "QV_Value": row[col],
                "Percentage": row[percent_col] if percent_col in df.columns else np.nan
            })

data = pd.DataFrame(rows)

data["QV_Value"] = pd.to_numeric(data["QV_Value"], errors="coerce")
data["Percentage"] = pd.to_numeric(data["Percentage"], errors="coerce")

data.dropna(inplace=True)

# ==============================
# ENCODING
# ==============================

market_encoder = LabelEncoder()
parameter_encoder = LabelEncoder()

data["Market_Encoded"] = market_encoder.fit_transform(data["Market"])
data["Parameter_Encoded"] = parameter_encoder.fit_transform(data["Parameter"])

joblib.dump(
    market_encoder,
    os.path.join(BASE_DIR, "qpso_market_encoder.pkl")
)

joblib.dump(
    parameter_encoder,
    os.path.join(BASE_DIR, "qpso_parameter_encoder.pkl")
)

# ==============================
# FEATURE ENGINEERING
# ==============================

data = data.sort_values(["Market", "Parameter", "Year"])

data["Lag_1"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].shift(1)

data["Lag_2"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].shift(2)

data["Rolling_Mean_3"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].transform(
    lambda x: x.rolling(3).mean()
)

data.dropna(inplace=True)

features = [
    "Year",
    "Percentage",
    "Market_Encoded",
    "Parameter_Encoded",
    "Lag_1",
    "Lag_2",
    "Rolling_Mean_3"
]

target = "QV_Value"

# ==============================
# TRAIN TEST SPLIT
# ==============================

train_data = data[data["Year"] <= 2011]
test_data = data[data["Year"] > 2011]

X_train = train_data[features].values
y_train = train_data[target].values

X_test = test_data[features].values
y_test = test_data[target].values

# ==============================
# FITNESS FUNCTION
# ==============================

def fitness_function(solution, X_train, y_train, X_test, y_test):

    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=6,
        random_state=42
    )

    model.fit(
        X_train[:, selected_features],
        y_train
    )

    predictions = model.predict(
        X_test[:, selected_features]
    )

    r2 = r2_score(y_test, predictions)

    feature_penalty = len(selected_features) / len(solution)

    return r2 - (0.01 * feature_penalty)

# ==============================
# QPSO FEATURE SELECTION
# ==============================

def qpso_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features,
    population_size=20,
    generations=30,
    beta=0.75
):

    positions = np.random.randint(
        0,
        2,
        size=(population_size, num_features)
    ).astype(float)

    personal_best_positions = positions.copy()

    personal_best_scores = np.array([
        fitness_function(
            p,
            X_train,
            y_train,
            X_test,
            y_test
        )
        for p in positions
    ])

    global_best_index = np.argmax(personal_best_scores)

    global_best_position = personal_best_positions[
        global_best_index
    ].copy()

    global_best_score = personal_best_scores[
        global_best_index
    ]

    fitness_history = []

    for generation in range(generations):

        mbest = np.mean(personal_best_positions, axis=0)

        for i in range(population_size):

            phi = np.random.rand(num_features)

            p = (
                phi * personal_best_positions[i]
                + (1 - phi) * global_best_position
            )

            u = np.random.rand(num_features)

            direction = np.where(
                np.random.rand(num_features) < 0.5,
                -1,
                1
            )

            new_position = (
                p
                + direction
                * beta
                * np.abs(mbest - positions[i])
                * np.log(1 / u)
            )

            binary_position = np.where(
                new_position > 0.5,
                1,
                0
            )

            score = fitness_function(
                binary_position,
                X_train,
                y_train,
                X_test,
                y_test
            )

            positions[i] = binary_position

            if score > personal_best_scores[i]:

                personal_best_scores[i] = score
                personal_best_positions[i] = binary_position.copy()

        best_index = np.argmax(personal_best_scores)

        if personal_best_scores[best_index] > global_best_score:

            global_best_score = personal_best_scores[best_index]

            global_best_position = personal_best_positions[
                best_index
            ].copy()

        fitness_history.append(global_best_score)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best Fitness: {global_best_score:.4f}"
        )

    return global_best_position, global_best_score, fitness_history

# ==============================
# RUN QPSO
# ==============================

best_solution, best_fitness, fitness_history = qpso_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features=len(features),
    population_size=20,
    generations=30,
    beta=0.75
)

selected_indices = np.where(best_solution == 1)[0]

selected_features = [
    features[i]
    for i in selected_indices
]

print("\nQPSO Selected Features:")
print(selected_features)

# ==============================
# FINAL MODEL
# ==============================

final_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    random_state=42
)

final_model.fit(
    X_train[:, selected_indices],
    y_train
)

predictions = final_model.predict(
    X_test[:, selected_indices]
)

# ==============================
# METRICS
# ==============================

mae = mean_absolute_error(y_test, predictions)

rmse = np.sqrt(
    mean_squared_error(y_test, predictions)
)

r2 = r2_score(y_test, predictions)

accuracy = r2 * 100

# ==============================
# SAVE MODEL
# ==============================

joblib.dump(
    final_model,
    os.path.join(BASE_DIR, "qpso_export_demand_model.pkl")
)

scaler = MinMaxScaler()

scaler.fit(data[[target]])

joblib.dump(
    scaler,
    os.path.join(BASE_DIR, "qpso_export_demand_scaler.pkl")
)

# ==============================
# RESULT CSV
# ==============================

result_csv_path = os.path.join(
    BASE_DIR,
    "qpso_export_demand_result.csv"
)

data.to_csv(result_csv_path, index=False)

# ==============================
# PREDICTION CSV
# ==============================

prediction_csv = test_data.copy()

prediction_csv["Actual_QV_Value"] = y_test

prediction_csv["Predicted_QV_Value"] = predictions

prediction_csv["Error"] = (
    prediction_csv["Actual_QV_Value"]
    - prediction_csv["Predicted_QV_Value"]
)

prediction_csv["Absolute_Error"] = abs(
    prediction_csv["Error"]
)

prediction_csv_path = os.path.join(
    BASE_DIR,
    "qpso_export_demand_prediction.csv"
)

prediction_csv.to_csv(
    prediction_csv_path,
    index=False
)

# ==============================
# SAVE JSON
# ==============================

metrics = {
    "Algorithm": "Quantum Particle Swarm Optimization + Random Forest",
    "Selected_Features": selected_features,
    "Best_QPSO_Fitness": float(best_fitness),
    "MAE": float(mae),
    "RMSE": float(rmse),
    "R2_Score": float(r2),
    "Accuracy_Percentage": float(accuracy)
}

json_path = os.path.join(
    BASE_DIR,
    "qpso_export_demand_metrics.json"
)

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# ==============================
# SAVE YAML
# ==============================

config = {
    "project_name": "Export Demand Forecasting System",
    "algorithm": "QPSO",
    "dataset_path": CSV_PATH,
    "output_directory": BASE_DIR,
    "target_column": target,
    "all_features": features,
    "selected_features": selected_features,
    "model": "RandomForestRegressor",
    "qpso_parameters": {
        "population_size": 20,
        "generations": 30,
        "beta": 0.75
    }
}

yaml_path = os.path.join(
    BASE_DIR,
    "qpso_export_demand_config.yaml"
)

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==============================
# RESULT GRAPH
# ==============================

yearly_result = data.groupby(
    "Year"
)["QV_Value"].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    yearly_result["Year"],
    yearly_result["QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Total Q/V Value")

plt.title("QPSO Result Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_result_graph.png")
)

plt.show()

# ==============================
# PREDICTION GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Predicted Q/V Value")

plt.title("QPSO Prediction Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_prediction_graph.png")
)

plt.show()

# ==============================
# COMPARISON GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Actual_QV_Value"],
    marker="o",
    label="Actual"
)

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="x",
    label="Predicted"
)

plt.xlabel("Year")
plt.ylabel("Q/V Value")

plt.title("QPSO Comparison Graph")

plt.legend()

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_comparison_graph.png")
)

plt.show()

# ==============================
# HEATMAP
# ==============================

heatmap_data = data.pivot_table(
    values="QV_Value",
    index="Market",
    columns="Year",
    aggfunc="sum"
)

plt.figure(figsize=(12, 7))

plt.imshow(heatmap_data, aspect="auto")

plt.colorbar(label="Q/V Value")

plt.xticks(
    range(len(heatmap_data.columns)),
    heatmap_data.columns,
    rotation=45
)

plt.yticks(
    range(len(heatmap_data.index)),
    heatmap_data.index
)

plt.title("QPSO Heatmap")

plt.xlabel("Year")
plt.ylabel("Market")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_heatmap.png")
)

plt.show()

# ==============================
# ACCURACY GRAPH
# ==============================

metric_names = [
    "MAE",
    "RMSE",
    "R2 Score",
    "Accuracy %"
]

metric_values = [
    mae,
    rmse,
    r2,
    accuracy
]

plt.figure(figsize=(10, 6))

plt.bar(metric_names, metric_values)

plt.xlabel("Metrics")
plt.ylabel("Values")

plt.title("QPSO Accuracy Graph")

plt.grid(axis="y")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_accuracy_graph.png")
)

plt.show()

# ==============================
# FITNESS GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    range(1, len(fitness_history) + 1),
    fitness_history,
    marker="o"
)

plt.xlabel("Generation")
plt.ylabel("Best Fitness")

plt.title("QPSO Fitness Convergence Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_fitness_graph.png")
)

plt.show()

# ==============================
# FEATURE IMPORTANCE GRAPH
# ==============================

feature_importance = pd.Series(
    final_model.feature_importances_,
    index=selected_features
).sort_values(ascending=False)

plt.figure(figsize=(10, 6))

feature_importance.plot(kind="bar")

plt.xlabel("Selected Features")
plt.ylabel("Importance")

plt.title("QPSO Feature Importance Graph")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "qpso_feature_importance_graph.png")
)

plt.show()

# ==============================
# FINAL OUTPUT
# ==============================

print("\nQPSO Project Execution Completed Successfully!")

print("\nSaved CSV Files:")
print(result_csv_path)
print(prediction_csv_path)

print("\nSaved Model:")
print(
    os.path.join(
        BASE_DIR,
        "qpso_export_demand_model.pkl"
    )
)

print("\nSaved Config Files:")
print(json_path)
print(yaml_path)

print("\nSaved Graphs:")
print("qpso_result_graph.png")
print("qpso_prediction_graph.png")
print("qpso_comparison_graph.png")
print("qpso_heatmap.png")
print("qpso_accuracy_graph.png")
print("qpso_fitness_graph.png")
print("qpso_feature_importance_graph.png")

print("\nQPSO Selected Features:")
print(selected_features)

print("\nMetrics:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
print("Accuracy %:", accuracy)
```
