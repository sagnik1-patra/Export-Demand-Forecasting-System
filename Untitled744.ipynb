```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ==============================
# PATHS
# ==============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Export Demand Forecasting System"
CSV_PATH = os.path.join(BASE_DIR, "Table_B-4.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv(CSV_PATH)

print("Original Dataset:")
print(df.head())

# ==============================
# WIDE TO LONG FORMAT
# ==============================

rows = []

for _, row in df.iterrows():
    market = row["Market"]
    parameter = row["Parameter"]

    for col in df.columns:
        if "Q/V" in col:
            year_label = col.split(" - ")[0]
            percent_col = f"{year_label} - %"

            rows.append({
                "Market": market,
                "Parameter": parameter,
                "Year_Label": year_label,
                "Year": int(year_label.split("-")[0]),
                "QV_Value": row[col],
                "Percentage": row[percent_col]
            })

data = pd.DataFrame(rows)

data["QV_Value"] = pd.to_numeric(data["QV_Value"], errors="coerce")
data["Percentage"] = pd.to_numeric(data["Percentage"], errors="coerce")
data.dropna(inplace=True)

# ==============================
# ENCODING
# ==============================

market_encoder = LabelEncoder()
parameter_encoder = LabelEncoder()

data["Market_Encoded"] = market_encoder.fit_transform(data["Market"])
data["Parameter_Encoded"] = parameter_encoder.fit_transform(data["Parameter"])

joblib.dump(market_encoder, os.path.join(BASE_DIR, "market_encoder.pkl"))
joblib.dump(parameter_encoder, os.path.join(BASE_DIR, "parameter_encoder.pkl"))

# ==============================
# FEATURE ENGINEERING
# ==============================

data = data.sort_values(["Market", "Parameter", "Year"])

data["Lag_1"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(1)
data["Lag_2"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(2)
data["Rolling_Mean_3"] = data.groupby(["Market", "Parameter"])["QV_Value"].transform(
    lambda x: x.rolling(3).mean()
)

data.dropna(inplace=True)

features = [
    "Year",
    "Percentage",
    "Market_Encoded",
    "Parameter_Encoded",
    "Lag_1",
    "Lag_2",
    "Rolling_Mean_3"
]

target = "QV_Value"

# ==============================
# TRAIN TEST SPLIT
# ==============================

train_data = data[data["Year"] <= 2011]
test_data = data[data["Year"] > 2011]

X_train = train_data[features]
y_train = train_data[target]

X_test = test_data[features]
y_test = test_data[target]

# ==============================
# RANDOM FOREST MODEL
# ==============================

rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    random_state=42
)

rf_model.fit(X_train, y_train)

predictions = rf_model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)
accuracy = r2 * 100

# ==============================
# SAVE PKL MODEL
# ==============================

joblib.dump(rf_model, os.path.join(BASE_DIR, "export_demand_rf_model.pkl"))

scaler = MinMaxScaler()
scaler.fit(data[[target]])
joblib.dump(scaler, os.path.join(BASE_DIR, "export_demand_scaler.pkl"))

# ==============================
# LSTM MODEL
# ==============================

yearly_data = data.groupby("Year")["QV_Value"].sum().reset_index()

lstm_scaler = MinMaxScaler()
yearly_data["Scaled_Value"] = lstm_scaler.fit_transform(yearly_data[["QV_Value"]])

sequence_length = 3
values = yearly_data["Scaled_Value"].values

X_lstm = []
y_lstm = []

for i in range(sequence_length, len(values)):
    X_lstm.append(values[i-sequence_length:i])
    y_lstm.append(values[i])

X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)

X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

lstm_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(sequence_length, 1)),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(1)
])

lstm_model.compile(optimizer="adam", loss="mse")

history = lstm_model.fit(
    X_lstm,
    y_lstm,
    epochs=200,
    batch_size=2,
    verbose=1,
    callbacks=[
        EarlyStopping(
            monitor="loss",
            patience=20,
            restore_best_weights=True
        )
    ]
)

lstm_model.save(os.path.join(BASE_DIR, "export_demand_lstm_model.h5"))
joblib.dump(lstm_scaler, os.path.join(BASE_DIR, "export_demand_lstm_scaler.pkl"))

# ==============================
# RESULT CSV
# ==============================

result_csv = data.copy()
result_csv_path = os.path.join(BASE_DIR, "export_demand_result.csv")
result_csv.to_csv(result_csv_path, index=False)

# ==============================
# PREDICTION CSV
# ==============================

prediction_csv = test_data.copy()
prediction_csv["Predicted_QV_Value"] = predictions
prediction_csv["Actual_QV_Value"] = y_test.values
prediction_csv["Error"] = prediction_csv["Actual_QV_Value"] - prediction_csv["Predicted_QV_Value"]
prediction_csv["Absolute_Error"] = abs(prediction_csv["Error"])

prediction_csv_path = os.path.join(BASE_DIR, "export_demand_prediction.csv")
prediction_csv.to_csv(prediction_csv_path, index=False)

# ==============================
# METRICS JSON
# ==============================

metrics = {
    "MAE": float(mae),
    "RMSE": float(rmse),
    "R2_Score": float(r2),
    "Accuracy_Percentage": float(accuracy),
    "LSTM_Final_Loss": float(history.history["loss"][-1])
}

with open(os.path.join(BASE_DIR, "export_demand_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# ==============================
# YAML CONFIG
# ==============================

config = {
    "project_name": "Export Demand Forecasting System",
    "dataset_path": CSV_PATH,
    "output_directory": BASE_DIR,
    "target_column": target,
    "features_used": features,
    "models_used": [
        "RandomForestRegressor",
        "LSTM"
    ],
    "saved_outputs": [
        "export_demand_result.csv",
        "export_demand_prediction.csv",
        "export_demand_rf_model.pkl",
        "export_demand_lstm_model.h5",
        "export_demand_metrics.json"
    ]
}

with open(os.path.join(BASE_DIR, "export_demand_config.yaml"), "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==============================
# GRAPH 1: RESULT GRAPH
# ==============================

yearly_result = data.groupby("Year")["QV_Value"].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(yearly_result["Year"], yearly_result["QV_Value"], marker="o")
plt.xlabel("Year")
plt.ylabel("Total Q/V Value")
plt.title("Result Graph - Year-wise Export Demand")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_result_graph.png"))
plt.show()

# ==============================
# GRAPH 2: PREDICTION GRAPH
# ==============================

plt.figure(figsize=(10, 6))
plt.plot(prediction_csv["Year"], prediction_csv["Predicted_QV_Value"], marker="o")
plt.xlabel("Year")
plt.ylabel("Predicted Q/V Value")
plt.title("Prediction Graph - Predicted Export Demand")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_prediction_graph.png"))
plt.show()

# ==============================
# GRAPH 3: COMPARISON GRAPH
# ==============================

plt.figure(figsize=(10, 6))
plt.plot(prediction_csv["Year"], prediction_csv["Actual_QV_Value"], marker="o", label="Actual")
plt.plot(prediction_csv["Year"], prediction_csv["Predicted_QV_Value"], marker="x", label="Predicted")
plt.xlabel("Year")
plt.ylabel("Q/V Value")
plt.title("Comparison Graph - Actual vs Predicted")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_comparison_graph.png"))
plt.show()

# ==============================
# GRAPH 4: HEATMAP
# ==============================

heatmap_data = data.pivot_table(
    values="QV_Value",
    index="Market",
    columns="Year",
    aggfunc="sum"
)

plt.figure(figsize=(12, 7))
plt.imshow(heatmap_data, aspect="auto")
plt.colorbar(label="Q/V Value")
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=45)
plt.yticks(range(len(heatmap_data.index)), heatmap_data.index)
plt.title("Heatmap - Market-wise Export Demand")
plt.xlabel("Year")
plt.ylabel("Market")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_heatmap.png"))
plt.show()

# ==============================
# GRAPH 5: ACCURACY GRAPH
# ==============================

metric_names = ["MAE", "RMSE", "R2 Score", "Accuracy %"]
metric_values = [mae, rmse, r2, accuracy]

plt.figure(figsize=(10, 6))
plt.bar(metric_names, metric_values)
plt.xlabel("Metrics")
plt.ylabel("Values")
plt.title("Accuracy Graph - Model Performance")
plt.grid(axis="y")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_accuracy_graph.png"))
plt.show()

# ==============================
# GRAPH 6: LSTM LOSS GRAPH
# ==============================

plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("LSTM Accuracy/Loss Graph")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_lstm_loss_graph.png"))
plt.show()

# ==============================
# GRAPH 7: FEATURE IMPORTANCE GRAPH
# ==============================

feature_importance = pd.Series(
    rf_model.feature_importances_,
    index=features
).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
feature_importance.plot(kind="bar")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance Graph")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "export_demand_feature_importance_graph.png"))
plt.show()

# ==============================
# FINAL OUTPUT
# ==============================

print("\nProject Execution Completed Successfully!")

print("\nSaved CSV Files:")
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)

print("\nSaved Models:")
print("PKL Model:", os.path.join(BASE_DIR, "export_demand_rf_model.pkl"))
print("H5 Model:", os.path.join(BASE_DIR, "export_demand_lstm_model.h5"))

print("\nSaved Config Files:")
print("YAML:", os.path.join(BASE_DIR, "export_demand_config.yaml"))
print("JSON:", os.path.join(BASE_DIR, "export_demand_metrics.json"))

print("\nSaved Graphs:")
print("export_demand_result_graph.png")
print("export_demand_prediction_graph.png")
print("export_demand_comparison_graph.png")
print("export_demand_heatmap.png")
print("export_demand_accuracy_graph.png")
print("export_demand_lstm_loss_graph.png")
print("export_demand_feature_importance_graph.png")

print("\nModel Metrics:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
print("Accuracy %:", accuracy)
```
