```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ==============================
# PATHS
# ==============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Export Demand Forecasting System"
CSV_PATH = os.path.join(BASE_DIR, "Table_B-4.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv(CSV_PATH)

# ==============================
# WIDE TO LONG FORMAT
# ==============================

rows = []

for _, row in df.iterrows():

    market = row["Market"]
    parameter = row["Parameter"]

    for col in df.columns:

        if "Q/V" in col:

            year_label = col.split(" - ")[0]
            percent_col = f"{year_label} - %"

            rows.append({
                "Market": market,
                "Parameter": parameter,
                "Year_Label": year_label,
                "Year": int(year_label.split("-")[0]),
                "QV_Value": row[col],
                "Percentage": row[percent_col] if percent_col in df.columns else np.nan
            })

data = pd.DataFrame(rows)

data["QV_Value"] = pd.to_numeric(data["QV_Value"], errors="coerce")
data["Percentage"] = pd.to_numeric(data["Percentage"], errors="coerce")

data.dropna(inplace=True)

# ==============================
# ENCODING
# ==============================

market_encoder = LabelEncoder()
parameter_encoder = LabelEncoder()

data["Market_Encoded"] = market_encoder.fit_transform(data["Market"])
data["Parameter_Encoded"] = parameter_encoder.fit_transform(data["Parameter"])

joblib.dump(
    market_encoder,
    os.path.join(BASE_DIR, "ba_market_encoder.pkl")
)

joblib.dump(
    parameter_encoder,
    os.path.join(BASE_DIR, "ba_parameter_encoder.pkl")
)

# ==============================
# FEATURE ENGINEERING
# ==============================

data = data.sort_values(["Market", "Parameter", "Year"])

data["Lag_1"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].shift(1)

data["Lag_2"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].shift(2)

data["Rolling_Mean_3"] = data.groupby(
    ["Market", "Parameter"]
)["QV_Value"].transform(
    lambda x: x.rolling(3).mean()
)

data.dropna(inplace=True)

features = [
    "Year",
    "Percentage",
    "Market_Encoded",
    "Parameter_Encoded",
    "Lag_1",
    "Lag_2",
    "Rolling_Mean_3"
]

target = "QV_Value"

# ==============================
# TRAIN TEST SPLIT
# ==============================

train_data = data[data["Year"] <= 2011]
test_data = data[data["Year"] > 2011]

X_train = train_data[features].values
y_train = train_data[target].values

X_test = test_data[features].values
y_test = test_data[target].values

# ==============================
# FITNESS FUNCTION
# ==============================

def fitness_function(solution, X_train, y_train, X_test, y_test):

    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=6,
        random_state=42
    )

    model.fit(
        X_train[:, selected_features],
        y_train
    )

    predictions = model.predict(
        X_test[:, selected_features]
    )

    r2 = r2_score(y_test, predictions)

    feature_penalty = len(selected_features) / len(solution)

    return r2 - (0.01 * feature_penalty)

# ==============================
# BAT ALGORITHM FEATURE SELECTION
# ==============================

def bat_algorithm_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features,
    population_size=20,
    generations=30,
    fmin=0,
    fmax=2,
    alpha=0.9,
    gamma=0.9
):

    bats = np.random.randint(
        0,
        2,
        size=(population_size, num_features)
    ).astype(float)

    velocities = np.zeros((population_size, num_features))
    frequencies = np.zeros(population_size)

    loudness = np.ones(population_size)
    pulse_rate = np.random.rand(population_size)

    fitness_scores = np.array([
        fitness_function(
            bat,
            X_train,
            y_train,
            X_test,
            y_test
        )
        for bat in bats
    ])

    best_index = np.argmax(fitness_scores)

    best_solution = bats[best_index].copy()
    best_fitness = fitness_scores[best_index]

    fitness_history = []

    for generation in range(generations):

        for i in range(population_size):

            frequencies[i] = fmin + (fmax - fmin) * np.random.rand()

            velocities[i] = (
                velocities[i]
                + (bats[i] - best_solution)
                * frequencies[i]
            )

            new_solution = bats[i] + velocities[i]

            new_solution = np.where(
                new_solution > 0.5,
                1,
                0
            )

            if np.random.rand() > pulse_rate[i]:

                epsilon = np.random.uniform(
                    -1,
                    1,
                    size=num_features
                )

                new_solution = (
                    best_solution
                    + epsilon * np.mean(loudness)
                )

                new_solution = np.where(
                    new_solution > 0.5,
                    1,
                    0
                )

            new_fitness = fitness_function(
                new_solution,
                X_train,
                y_train,
                X_test,
                y_test
            )

            if (
                new_fitness > fitness_scores[i]
                and np.random.rand() < loudness[i]
            ):

                bats[i] = new_solution.copy()
                fitness_scores[i] = new_fitness

                loudness[i] = alpha * loudness[i]

                pulse_rate[i] = (
                    pulse_rate[i]
                    * (1 - np.exp(-gamma * generation))
                )

            if new_fitness > best_fitness:

                best_solution = new_solution.copy()
                best_fitness = new_fitness

        fitness_history.append(best_fitness)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best Fitness: {best_fitness:.4f}"
        )

    return best_solution, best_fitness, fitness_history

# ==============================
# RUN BAT ALGORITHM
# ==============================

best_solution, best_fitness, fitness_history = bat_algorithm_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features=len(features),
    population_size=20,
    generations=30,
    fmin=0,
    fmax=2,
    alpha=0.9,
    gamma=0.9
)

selected_indices = np.where(best_solution == 1)[0]

selected_features = [
    features[i]
    for i in selected_indices
]

print("\nBA Selected Features:")
print(selected_features)

# ==============================
# FINAL MODEL
# ==============================

final_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    random_state=42
)

final_model.fit(
    X_train[:, selected_indices],
    y_train
)

predictions = final_model.predict(
    X_test[:, selected_indices]
)

# ==============================
# METRICS
# ==============================

mae = mean_absolute_error(y_test, predictions)

rmse = np.sqrt(
    mean_squared_error(y_test, predictions)
)

r2 = r2_score(y_test, predictions)

accuracy = r2 * 100

# ==============================
# SAVE MODEL
# ==============================

joblib.dump(
    final_model,
    os.path.join(BASE_DIR, "ba_export_demand_model.pkl")
)

scaler = MinMaxScaler()

scaler.fit(data[[target]])

joblib.dump(
    scaler,
    os.path.join(BASE_DIR, "ba_export_demand_scaler.pkl")
)

# ==============================
# RESULT CSV
# ==============================

result_csv_path = os.path.join(
    BASE_DIR,
    "ba_export_demand_result.csv"
)

data.to_csv(result_csv_path, index=False)

# ==============================
# PREDICTION CSV
# ==============================

prediction_csv = test_data.copy()

prediction_csv["Actual_QV_Value"] = y_test

prediction_csv["Predicted_QV_Value"] = predictions

prediction_csv["Error"] = (
    prediction_csv["Actual_QV_Value"]
    - prediction_csv["Predicted_QV_Value"]
)

prediction_csv["Absolute_Error"] = abs(
    prediction_csv["Error"]
)

prediction_csv_path = os.path.join(
    BASE_DIR,
    "ba_export_demand_prediction.csv"
)

prediction_csv.to_csv(
    prediction_csv_path,
    index=False
)

# ==============================
# SAVE JSON
# ==============================

metrics = {
    "Algorithm": "Bat Algorithm + Random Forest",
    "Selected_Features": selected_features,
    "Best_BA_Fitness": float(best_fitness),
    "MAE": float(mae),
    "RMSE": float(rmse),
    "R2_Score": float(r2),
    "Accuracy_Percentage": float(accuracy)
}

json_path = os.path.join(
    BASE_DIR,
    "ba_export_demand_metrics.json"
)

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# ==============================
# SAVE YAML
# ==============================

config = {
    "project_name": "Export Demand Forecasting System",
    "algorithm": "BA",
    "dataset_path": CSV_PATH,
    "output_directory": BASE_DIR,
    "target_column": target,
    "all_features": features,
    "selected_features": selected_features,
    "model": "RandomForestRegressor",
    "ba_parameters": {
        "population_size": 20,
        "generations": 30,
        "fmin": 0,
        "fmax": 2,
        "alpha": 0.9,
        "gamma": 0.9
    }
}

yaml_path = os.path.join(
    BASE_DIR,
    "ba_export_demand_config.yaml"
)

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==============================
# RESULT GRAPH
# ==============================

yearly_result = data.groupby(
    "Year"
)["QV_Value"].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    yearly_result["Year"],
    yearly_result["QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Total Q/V Value")

plt.title("BA Result Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_result_graph.png")
)

plt.show()

# ==============================
# PREDICTION GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Predicted Q/V Value")

plt.title("BA Prediction Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_prediction_graph.png")
)

plt.show()

# ==============================
# COMPARISON GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Actual_QV_Value"],
    marker="o",
    label="Actual"
)

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="x",
    label="Predicted"
)

plt.xlabel("Year")
plt.ylabel("Q/V Value")

plt.title("BA Comparison Graph")

plt.legend()

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_comparison_graph.png")
)

plt.show()

# ==============================
# HEATMAP
# ==============================

heatmap_data = data.pivot_table(
    values="QV_Value",
    index="Market",
    columns="Year",
    aggfunc="sum"
)

plt.figure(figsize=(12, 7))

plt.imshow(heatmap_data, aspect="auto")

plt.colorbar(label="Q/V Value")

plt.xticks(
    range(len(heatmap_data.columns)),
    heatmap_data.columns,
    rotation=45
)

plt.yticks(
    range(len(heatmap_data.index)),
    heatmap_data.index
)

plt.title("BA Heatmap")

plt.xlabel("Year")
plt.ylabel("Market")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_heatmap.png")
)

plt.show()

# ==============================
# ACCURACY GRAPH
# ==============================

metric_names = [
    "MAE",
    "RMSE",
    "R2 Score",
    "Accuracy %"
]

metric_values = [
    mae,
    rmse,
    r2,
    accuracy
]

plt.figure(figsize=(10, 6))

plt.bar(metric_names, metric_values)

plt.xlabel("Metrics")
plt.ylabel("Values")

plt.title("BA Accuracy Graph")

plt.grid(axis="y")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_accuracy_graph.png")
)

plt.show()

# ==============================
# FITNESS GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    range(1, len(fitness_history) + 1),
    fitness_history,
    marker="o"
)

plt.xlabel("Generation")
plt.ylabel("Best Fitness")

plt.title("BA Fitness Convergence Graph")

plt.grid(True)

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_fitness_graph.png")
)

plt.show()

# ==============================
# FEATURE IMPORTANCE GRAPH
# ==============================

feature_importance = pd.Series(
    final_model.feature_importances_,
    index=selected_features
).sort_values(ascending=False)

plt.figure(figsize=(10, 6))

feature_importance.plot(kind="bar")

plt.xlabel("Selected Features")
plt.ylabel("Importance")

plt.title("BA Feature Importance Graph")

plt.tight_layout()

plt.savefig(
    os.path.join(BASE_DIR, "ba_feature_importance_graph.png")
)

plt.show()

# ==============================
# FINAL OUTPUT
# ==============================

print("\nBA Project Execution Completed Successfully!")

print("\nSaved CSV Files:")
print(result_csv_path)
print(prediction_csv_path)

print("\nSaved Model:")
print(
    os.path.join(
        BASE_DIR,
        "ba_export_demand_model.pkl"
    )
)

print("\nSaved Config Files:")
print(json_path)
print(yaml_path)

print("\nSaved Graphs:")
print("ba_result_graph.png")
print("ba_prediction_graph.png")
print("ba_comparison_graph.png")
print("ba_heatmap.png")
print("ba_accuracy_graph.png")
print("ba_fitness_graph.png")
print("ba_feature_importance_graph.png")

print("\nBA Selected Features:")
print(selected_features)

print("\nMetrics:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
print("Accuracy %:", accuracy)
```
