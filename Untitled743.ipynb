```python
# ==============================
# Export Demand Forecasting System
# Author: Sagnik Patra
# ==============================

import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ==============================
# PATHS
# ==============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Export Demand Forecasting System"
CSV_PATH = os.path.join(BASE_DIR, "Table_B-4.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv(CSV_PATH)

print("Original Data Shape:", df.shape)
print(df.head())

# ==============================
# CONVERT WIDE TO LONG FORMAT
# ==============================

long_rows = []

for _, row in df.iterrows():
    market = row["Market"]
    parameter = row["Parameter"]

    for col in df.columns:
        if "Q/V" in col:
            year = col.split(" - ")[0]
            percent_col = f"{year} - %"

            long_rows.append({
                "Market": market,
                "Parameter": parameter,
                "Year_Label": year,
                "Year": int(year.split("-")[0]),
                "QV_Value": row[col],
                "Percentage": row[percent_col] if percent_col in df.columns else np.nan
            })

data = pd.DataFrame(long_rows)

data["QV_Value"] = pd.to_numeric(data["QV_Value"], errors="coerce")
data["Percentage"] = pd.to_numeric(data["Percentage"], errors="coerce")

data.dropna(inplace=True)

# Save cleaned data
cleaned_path = os.path.join(BASE_DIR, "cleaned_export_demand_data.csv")
data.to_csv(cleaned_path, index=False)

print("Cleaned Data Shape:", data.shape)

# ==============================
# ENCODING
# ==============================

market_encoder = LabelEncoder()
parameter_encoder = LabelEncoder()

data["Market_Encoded"] = market_encoder.fit_transform(data["Market"])
data["Parameter_Encoded"] = parameter_encoder.fit_transform(data["Parameter"])

joblib.dump(market_encoder, os.path.join(BASE_DIR, "market_encoder.pkl"))
joblib.dump(parameter_encoder, os.path.join(BASE_DIR, "parameter_encoder.pkl"))

# ==============================
# FEATURE ENGINEERING
# ==============================

data = data.sort_values(["Market", "Parameter", "Year"])

data["Lag_1"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(1)
data["Lag_2"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(2)
data["Rolling_Mean_3"] = data.groupby(["Market", "Parameter"])["QV_Value"].transform(
    lambda x: x.rolling(3).mean()
)

data.dropna(inplace=True)

features = [
    "Year",
    "Percentage",
    "Market_Encoded",
    "Parameter_Encoded",
    "Lag_1",
    "Lag_2",
    "Rolling_Mean_3"
]

target = "QV_Value"

X = data[features]
y = data[target]

# ==============================
# TRAIN TEST SPLIT BY YEAR
# ==============================

train_data = data[data["Year"] <= 2011]
test_data = data[data["Year"] > 2011]

X_train = train_data[features]
y_train = train_data[target]

X_test = test_data[features]
y_test = test_data[target]

# ==============================
# RANDOM FOREST MODEL
# ==============================

rf_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    max_depth=8
)

rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

rf_mae = mean_absolute_error(y_test, rf_predictions)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
rf_r2 = r2_score(y_test, rf_predictions)

# Save PKL model
rf_model_path = os.path.join(BASE_DIR, "export_demand_rf_model.pkl")
joblib.dump(rf_model, rf_model_path)

# Save scaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(data[[target]])
joblib.dump(scaler, os.path.join(BASE_DIR, "export_demand_scaler.pkl"))

# ==============================
# LSTM MODEL
# ==============================

yearly_data = data.groupby("Year")["QV_Value"].sum().reset_index()

scaler_lstm = MinMaxScaler()
yearly_data["Scaled_Value"] = scaler_lstm.fit_transform(yearly_data[["QV_Value"]])

sequence_length = 3

X_lstm = []
y_lstm = []

values = yearly_data["Scaled_Value"].values

for i in range(sequence_length, len(values)):
    X_lstm.append(values[i-sequence_length:i])
    y_lstm.append(values[i])

X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)

X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

lstm_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(sequence_length, 1)),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(1)
])

lstm_model.compile(optimizer="adam", loss="mse")

early_stop = EarlyStopping(
    monitor="loss",
    patience=20,
    restore_best_weights=True
)

history = lstm_model.fit(
    X_lstm,
    y_lstm,
    epochs=200,
    batch_size=2,
    verbose=1,
    callbacks=[early_stop]
)

# Save H5 model
h5_model_path = os.path.join(BASE_DIR, "export_demand_lstm_model.h5")
lstm_model.save(h5_model_path)

joblib.dump(scaler_lstm, os.path.join(BASE_DIR, "export_demand_lstm_scaler.pkl"))

# ==============================
# PREDICTIONS CSV
# ==============================

results = test_data.copy()
results["RF_Predicted_QV_Value"] = rf_predictions
results["Error"] = results["QV_Value"] - results["RF_Predicted_QV_Value"]

prediction_path = os.path.join(BASE_DIR, "export_demand_predictions.csv")
results.to_csv(prediction_path, index=False)

# ==============================
# SAVE METRICS JSON
# ==============================

metrics = {
    "Random_Forest": {
        "MAE": float(rf_mae),
        "RMSE": float(rf_rmse),
        "R2_Score": float(rf_r2)
    },
    "LSTM": {
        "Final_Training_Loss": float(history.history["loss"][-1])
    },
    "Files_Saved": {
        "H5_Model": h5_model_path,
        "PKL_Model": rf_model_path,
        "Cleaned_Data": cleaned_path,
        "Predictions": prediction_path
    }
}

json_path = os.path.join(BASE_DIR, "export_demand_metrics.json")

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# ==============================
# SAVE CONFIG YAML
# ==============================

config = {
    "project_name": "Export Demand Forecasting System",
    "dataset_path": CSV_PATH,
    "output_directory": BASE_DIR,
    "target_column": target,
    "features_used": features,
    "models": {
        "RandomForestRegressor": {
            "n_estimators": 300,
            "max_depth": 8,
            "random_state": 42
        },
        "LSTM": {
            "sequence_length": sequence_length,
            "epochs": 200,
            "batch_size": 2,
            "optimizer": "adam",
            "loss": "mse"
        }
    }
}

yaml_path = os.path.join(BASE_DIR, "export_demand_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==============================
# GRAPHS
# ==============================

plt.figure(figsize=(10, 6))
plt.plot(results["Year"], results["QV_Value"], marker="o", label="Actual")
plt.plot(results["Year"], results["RF_Predicted_QV_Value"], marker="x", label="Predicted")
plt.xlabel("Year")
plt.ylabel("Export Demand Q/V Value")
plt.title("Actual vs Predicted Export Demand")
plt.legend()
plt.grid(True)

graph_path = os.path.join(BASE_DIR, "export_demand_actual_vs_predicted.png")
plt.savefig(graph_path)
plt.close()

plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("LSTM Training Loss")
plt.grid(True)

loss_graph_path = os.path.join(BASE_DIR, "export_demand_lstm_loss_graph.png")
plt.savefig(loss_graph_path)
plt.close()

plt.figure(figsize=(10, 6))
feature_importance = pd.Series(
    rf_model.feature_importances_,
    index=features
).sort_values(ascending=False)

feature_importance.plot(kind="bar")
plt.title("Feature Importance - Random Forest")
plt.ylabel("Importance")
plt.tight_layout()

importance_graph_path = os.path.join(BASE_DIR, "export_demand_feature_importance.png")
plt.savefig(importance_graph_path)
plt.close()

# ==============================
# SUMMARY
# ==============================

print("\nTraining Completed Successfully!")
print("Saved Files:")
print("H5 Model:", h5_model_path)
print("PKL Model:", rf_model_path)
print("YAML Config:", yaml_path)
print("JSON Metrics:", json_path)
print("Prediction CSV:", prediction_path)
print("Graphs Saved Successfully!")
```
