```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ==============================
# PATHS
# ==============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Export Demand Forecasting System"
CSV_PATH = os.path.join(BASE_DIR, "Table_B-4.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv(CSV_PATH)

# ==============================
# WIDE TO LONG FORMAT
# ==============================

rows = []

for _, row in df.iterrows():
    market = row["Market"]
    parameter = row["Parameter"]

    for col in df.columns:
        if "Q/V" in col:
            year_label = col.split(" - ")[0]
            percent_col = f"{year_label} - %"

            rows.append({
                "Market": market,
                "Parameter": parameter,
                "Year_Label": year_label,
                "Year": int(year_label.split("-")[0]),
                "QV_Value": row[col],
                "Percentage": row[percent_col] if percent_col in df.columns else np.nan
            })

data = pd.DataFrame(rows)

data["QV_Value"] = pd.to_numeric(data["QV_Value"], errors="coerce")
data["Percentage"] = pd.to_numeric(data["Percentage"], errors="coerce")
data.dropna(inplace=True)

# ==============================
# ENCODING
# ==============================

market_encoder = LabelEncoder()
parameter_encoder = LabelEncoder()

data["Market_Encoded"] = market_encoder.fit_transform(data["Market"])
data["Parameter_Encoded"] = parameter_encoder.fit_transform(data["Parameter"])

joblib.dump(market_encoder, os.path.join(BASE_DIR, "csa_market_encoder.pkl"))
joblib.dump(parameter_encoder, os.path.join(BASE_DIR, "csa_parameter_encoder.pkl"))

# ==============================
# FEATURE ENGINEERING
# ==============================

data = data.sort_values(["Market", "Parameter", "Year"])

data["Lag_1"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(1)
data["Lag_2"] = data.groupby(["Market", "Parameter"])["QV_Value"].shift(2)
data["Rolling_Mean_3"] = data.groupby(["Market", "Parameter"])["QV_Value"].transform(
    lambda x: x.rolling(3).mean()
)

data.dropna(inplace=True)

features = [
    "Year",
    "Percentage",
    "Market_Encoded",
    "Parameter_Encoded",
    "Lag_1",
    "Lag_2",
    "Rolling_Mean_3"
]

target = "QV_Value"

# ==============================
# TRAIN TEST SPLIT
# ==============================

train_data = data[data["Year"] <= 2011]
test_data = data[data["Year"] > 2011]

X_train = train_data[features].values
y_train = train_data[target].values

X_test = test_data[features].values
y_test = test_data[target].values

# ==============================
# CSA FITNESS FUNCTION
# ==============================

def fitness_function(solution, X_train, y_train, X_test, y_test):

    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=6,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)

    predictions = model.predict(X_test[:, selected_features])

    r2 = r2_score(y_test, predictions)

    feature_penalty = len(selected_features) / len(solution)

    fitness = r2 - (0.01 * feature_penalty)

    return fitness

# ==============================
# CSA FEATURE SELECTION
# ==============================

def csa_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features,
    population_size=20,
    generations=30,
    awareness_probability=0.2,
    flight_length=2
):

    population = np.random.randint(0, 2, size=(population_size, num_features))

    best_solution = None
    best_fitness = -9999

    fitness_history = []

    for generation in range(generations):

        fitness_scores = np.array([
            fitness_function(individual, X_train, y_train, X_test, y_test)
            for individual in population
        ])

        best_index = np.argmax(fitness_scores)

        if fitness_scores[best_index] > best_fitness:
            best_fitness = fitness_scores[best_index]
            best_solution = population[best_index].copy()

        new_population = population.copy()

        for i in range(population_size):

            if np.random.rand() > awareness_probability:

                random_crow = np.random.randint(population_size)

                step = np.random.randint(
                    -flight_length,
                    flight_length + 1,
                    size=num_features
                )

                new_solution = (
                    population[i]
                    + step * (population[random_crow] - population[i])
                )

                new_solution = np.where(new_solution > 0.5, 1, 0)

            else:
                new_solution = np.random.randint(0, 2, size=num_features)

            new_population[i] = new_solution

        population = new_population

        fitness_history.append(best_fitness)

        print(f"Generation {generation + 1}/{generations} | Best Fitness: {best_fitness:.4f}")

    return best_solution, best_fitness, fitness_history

# ==============================
# RUN CSA
# ==============================

best_solution, best_fitness, fitness_history = csa_feature_selection(
    X_train,
    y_train,
    X_test,
    y_test,
    num_features=len(features),
    population_size=20,
    generations=30,
    awareness_probability=0.2,
    flight_length=2
)

selected_indices = np.where(best_solution == 1)[0]
selected_features = [features[i] for i in selected_indices]

print("\nCSA Selected Features:")
print(selected_features)

# ==============================
# FINAL MODEL
# ==============================

final_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    random_state=42
)

final_model.fit(X_train[:, selected_indices], y_train)

predictions = final_model.predict(X_test[:, selected_indices])

# ==============================
# METRICS
# ==============================

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)
accuracy = r2 * 100

# ==============================
# SAVE MODEL
# ==============================

joblib.dump(final_model, os.path.join(BASE_DIR, "csa_export_demand_model.pkl"))

scaler = MinMaxScaler()
scaler.fit(data[[target]])

joblib.dump(scaler, os.path.join(BASE_DIR, "csa_export_demand_scaler.pkl"))

# ==============================
# RESULT CSV
# ==============================

result_csv = data.copy()

result_csv_path = os.path.join(BASE_DIR, "csa_export_demand_result.csv")

result_csv.to_csv(result_csv_path, index=False)

# ==============================
# PREDICTION CSV
# ==============================

prediction_csv = test_data.copy()

prediction_csv["Actual_QV_Value"] = y_test
prediction_csv["Predicted_QV_Value"] = predictions

prediction_csv["Error"] = (
    prediction_csv["Actual_QV_Value"]
    - prediction_csv["Predicted_QV_Value"]
)

prediction_csv["Absolute_Error"] = abs(prediction_csv["Error"])

prediction_csv_path = os.path.join(BASE_DIR, "csa_export_demand_prediction.csv")

prediction_csv.to_csv(prediction_csv_path, index=False)

# ==============================
# SAVE JSON
# ==============================

metrics = {
    "Algorithm": "Crow Search Algorithm Feature Selection + Random Forest",
    "Selected_Features": selected_features,
    "Best_CSA_Fitness": float(best_fitness),
    "MAE": float(mae),
    "RMSE": float(rmse),
    "R2_Score": float(r2),
    "Accuracy_Percentage": float(accuracy)
}

json_path = os.path.join(BASE_DIR, "csa_export_demand_metrics.json")

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# ==============================
# SAVE YAML
# ==============================

config = {
    "project_name": "Export Demand Forecasting System",
    "algorithm": "CSA",
    "dataset_path": CSV_PATH,
    "output_directory": BASE_DIR,
    "target_column": target,
    "all_features": features,
    "selected_features": selected_features,
    "model": "RandomForestRegressor",
    "csa_parameters": {
        "population_size": 20,
        "generations": 30,
        "awareness_probability": 0.2,
        "flight_length": 2
    }
}

yaml_path = os.path.join(BASE_DIR, "csa_export_demand_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==============================
# GRAPH 1: RESULT GRAPH
# ==============================

yearly_result = data.groupby("Year")["QV_Value"].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    yearly_result["Year"],
    yearly_result["QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Total Q/V Value")
plt.title("CSA Result Graph")

plt.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_result_graph.png"))

plt.show()

# ==============================
# GRAPH 2: PREDICTION GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="o"
)

plt.xlabel("Year")
plt.ylabel("Predicted Q/V Value")
plt.title("CSA Prediction Graph")

plt.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_prediction_graph.png"))

plt.show()

# ==============================
# GRAPH 3: COMPARISON GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Actual_QV_Value"],
    marker="o",
    label="Actual"
)

plt.plot(
    prediction_csv["Year"],
    prediction_csv["Predicted_QV_Value"],
    marker="x",
    label="Predicted"
)

plt.xlabel("Year")
plt.ylabel("Q/V Value")

plt.title("CSA Comparison Graph")

plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_comparison_graph.png"))

plt.show()

# ==============================
# GRAPH 4: HEATMAP
# ==============================

heatmap_data = data.pivot_table(
    values="QV_Value",
    index="Market",
    columns="Year",
    aggfunc="sum"
)

plt.figure(figsize=(12, 7))

plt.imshow(heatmap_data, aspect="auto")

plt.colorbar(label="Q/V Value")

plt.xticks(
    range(len(heatmap_data.columns)),
    heatmap_data.columns,
    rotation=45
)

plt.yticks(
    range(len(heatmap_data.index)),
    heatmap_data.index
)

plt.title("CSA Heatmap")

plt.xlabel("Year")
plt.ylabel("Market")

plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_heatmap.png"))

plt.show()

# ==============================
# GRAPH 5: ACCURACY GRAPH
# ==============================

metric_names = ["MAE", "RMSE", "R2 Score", "Accuracy %"]
metric_values = [mae, rmse, r2, accuracy]

plt.figure(figsize=(10, 6))

plt.bar(metric_names, metric_values)

plt.xlabel("Metrics")
plt.ylabel("Values")

plt.title("CSA Accuracy Graph")

plt.grid(axis="y")
plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_accuracy_graph.png"))

plt.show()

# ==============================
# GRAPH 6: CSA FITNESS GRAPH
# ==============================

plt.figure(figsize=(10, 6))

plt.plot(
    range(1, len(fitness_history) + 1),
    fitness_history,
    marker="o"
)

plt.xlabel("Generation")
plt.ylabel("Best Fitness")

plt.title("CSA Fitness Convergence Graph")

plt.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_fitness_graph.png"))

plt.show()

# ==============================
# GRAPH 7: FEATURE IMPORTANCE
# ==============================

feature_importance = pd.Series(
    final_model.feature_importances_,
    index=selected_features
).sort_values(ascending=False)

plt.figure(figsize=(10, 6))

feature_importance.plot(kind="bar")

plt.xlabel("Selected Features")
plt.ylabel("Importance")

plt.title("CSA Feature Importance Graph")

plt.tight_layout()

plt.savefig(os.path.join(BASE_DIR, "csa_feature_importance_graph.png"))

plt.show()

# ==============================
# FINAL OUTPUT
# ==============================

print("\nCSA Project Execution Completed Successfully!")

print("\nSaved CSV Files:")
print(result_csv_path)
print(prediction_csv_path)

print("\nSaved Model:")
print(os.path.join(BASE_DIR, "csa_export_demand_model.pkl"))

print("\nSaved Config Files:")
print(json_path)
print(yaml_path)

print("\nSaved Graphs:")
print("csa_result_graph.png")
print("csa_prediction_graph.png")
print("csa_comparison_graph.png")
print("csa_heatmap.png")
print("csa_accuracy_graph.png")
print("csa_fitness_graph.png")
print("csa_feature_importance_graph.png")

print("\nCSA Selected Features:")
print(selected_features)

print("\nMetrics:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
print("Accuracy %:", accuracy)
```
